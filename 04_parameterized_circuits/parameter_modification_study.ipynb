# Parameterized Circuit Modification Study

This notebook studies how changes in ansatz parameters and circuit depth affect the circuit.

```python
from qiskit.circuit.library import RealAmplitudes
from qiskit.primitives import StatevectorEstimator
from qiskit.quantum_info import SparsePauliOp
```

```python
circuit = RealAmplitudes(num_qubits=2, entanglement="linear", reps=1)
print(circuit)
print("Parameters:", list(circuit.parameters))
```

```python
values = [3.14, 1.57, 0.50, 2.00]
bound = circuit.assign_parameters(values)
print(bound.decompose())
```

## Measure an expectation value

```python
observable = SparsePauliOp.from_list([("ZZ", 1.0)])
estimator = StatevectorEstimator()
result = estimator.run([(circuit, observable, [values])]).result()[0]
print("<ZZ> =", result.data.evs)
```
