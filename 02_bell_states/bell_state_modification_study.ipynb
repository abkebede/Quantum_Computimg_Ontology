# Bell-State Modification Study

Run the original circuit first. Then modify one line at a time and predict the result before running.

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
```

## Original Bell state

```python
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])
qc.draw("mpl")
```

```python
sampler = StatevectorSampler(seed=123)
result = sampler.run([qc], shots=1000).result()
counts = result[0].data.c.get_counts()
print(counts)
```

## Modification A: Change `qc.h(0)` to `qc.x(0)`

Prediction: The circuit no longer creates equal superposition before entanglement.

```python
qc = QuantumCircuit(2, 2)
qc.x(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])
result = sampler.run([qc], shots=1000).result()
print(result[0].data.c.get_counts())
```
